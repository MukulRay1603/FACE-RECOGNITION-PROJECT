<H2> GRAY SCALE CONVERSION

```python
limitx=100

jennie_gray=[None]*limit
j=0

for i in jennie:
    if(j<limitx):
        jennie_gray[j] = rgb2gray(jennie_images[j])
        j+=1
        
    else:  
        break
        
print('Image set - grayscale processed.')
```

<H2>Image length of gray scale images

```python
len_of_images_jen=len(jennie_gray)
```

```python
#variable image_size_selena
image_size_jen=jennie_gray[1].shape
image_size_jen
```

<H2> Flattening

```python
flatten_size_jennie=image_size_jen[0]*image_size_jen[1]
flatten_size_jennie
```

```python
#Flatten size of image containing product of 512x512
for i in range(len_of_images_jen):
    jennie_gray[i] = np.ndarray.flatten(jennie_gray[i]).reshape(flatten_size_jennie,1)
```

```python
#Stack individual image array elements into one array
jennie_gray = np.dstack(jennie_gray)
```

```python
#change the axis of array elements
jennie_gray = np.rollaxis(jennie_gray, axis=2, start=0)
jennie_gray.shape
```

```python
jennie_gray = jennie_gray.reshape(len_of_images_jen, flatten_size_jennie)
jennie_gray.shape
```

<H2> Concat dataset

```python
actor_1 = pd.concat((jenniedf,bendf))
actor_1
```

```python
actors = pd.concat((actor_1,selenadf))
actors
```

<H2>Removing index from final dataset

```python
from sklearn.utils import shuffle

shuff = shuffle(actors).reset_index()
shuff
```

```python
shuff = shuff.drop(["index"],axis=1)
#shuff = shuff.dropna()
shuff
```

<H2> Assign  Dependant and Independant Variables 

```python
x = shuff.values[:,:-1]
x
```

```python
y = shuff.values[:,-1]
y
```

<H2> TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
#SVM Classification
clf=svm.SVC()
clf.fit(x_train, y_train)
```

<H2> F1-Image Prediction

```python
y_pred=clf.predict(x_test)
y_pred
```

<H2> PREDICTION ACCURACY

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

<H2> Error Analysis Prediction

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
